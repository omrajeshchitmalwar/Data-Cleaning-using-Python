# Import Data

```python
data = pd.read_csv(r'C:\Users\91960\OneDrive\Desktop\Project\PYTHON\Data cleaning\Customer Call List.csv')
data
```

# Removing Duplicates

```python
data.duplicated()
```

```python
data1 = data.drop_duplicates()
data1
```

# Drop useless columns

```python
data1 = data1.drop(columns = "Unnamed: 9")
data1
```

```python
data1 = data1.drop(columns = "Unnamed: 8")
data1
```

```python
data1 = data1.drop(columns = "Not_Useful_Column")
data1
```

# Strip

```python
data1['Last_Name'] = data1['Last_Name'].str.lstrip("/")
data1['Last_Name'] = data1['Last_Name'].str.lstrip("...")
data1['Last_Name'] = data1['Last_Name'].str.rstrip("_")
#data1['Last_Name'] = data1['Last_Name'].str.strip("123._/")
data1
```

# Cleaning

```python
data1['Phone_Number'] = data1['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
data1
```

```python
data1['Phone_Number'] = data1['Phone_Number'].str.replace('Na',' ', regex = False)
data1
```

```python
data1['Phone_Number'] = data1['Phone_Number'].str.replace('NaN',' ', regex = False)
data1
```

```python
data1['Phone_Number'] = data1['Phone_Number'].fillna(' ') 
data1
```

```python
data1['Last_Name'] = data1['Last_Name'].fillna(' ') 
data1
```

```python
data1['Address'] = data1['Address'].str.replace('N/a',' ', regex = False)
data1
```

```python
data1[["Street_Address", "State", "Zip_Code"]] = data1["Address"].str.split(',',2, expand=True)
data1
```

```python
data1['Do_Not_Contact'] = data1['Do_Not_Contact'].str.replace('No','N', regex = False)
data1
```

```python
for x in data1.index:
    if data1.loc[x,'Do_Not_Contact'] == 'Y':
        data1.drop(x, inplace=True)
        
data1        
```

```python
data1 = data1.fillna('')
data1
```

```python
data1['Do_Not_Contact'] = data1['Do_Not_Contact'].str.replace('Yes', 'Y', regex=False)
data1
```

```python
data1['Paying Customer'] = data1['Paying Customer'].str.replace('Yes', 'Y', regex=False)
data1
```

```python
data1['Paying Customer'] = data1['Paying Customer'].str.replace('No', 'N', regex=False)
data1
```

```python
data1['Paying Customer'] = data1['Paying Customer'].str.replace('N/a', ' ', regex=False)
data1
```

```python
for x in data1.index:
    if data1.loc[x, "Phone_Number"] == ' ':
        data1.drop(x, inplace=True)

data1
```

```python
data1 = data1.reset_index(drop=True)
data1
```
